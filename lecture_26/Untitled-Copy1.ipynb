```python
X = np.linspace(0, 10, 100)
```

```python
m, c = 5, 50
```

```python
noise = np.random.randint(-10, 10, 100)
```

```python
y = m * X + c + noise
```

```python
plt.scatter(X, y)
```

```python
class LinearRegCustom:
    
    def __init__(self, lr=1, max_steps=10):
        self.lr = lr
        self.max_steps = max_steps
    
    def fit(self, X, y):
        self.coef_ = np.random.random()
        self.intercept_ = np.random.random()
        for i in range(self.max_steps):
            self.gradient_descend(X, y)
    
    def gradient_descend(self, X, y):
        delta_coef, delta_int = self.gradient(X, y)
#         print(delta_coef)
        self.coef_ += delta_coef
        self.intercept_ += delta_int
    def gradient(self, X, y):
        yp = self.predict(X)
        d_e2m = np.mean((yp - y) * X)
        d_e2c = np.mean(yp - y)
        return -d_e2m * self.lr, -d_e2c * self.lr
    
    def predict(self, X):
        return X * self.coef_ + self.intercept_
```

```python
model = LinearRegCustom(lr = 0.01, max_steps = 2000)
```

```python
model.fit(X, y)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(X, model.coef_*X + model.intercept_)
```
