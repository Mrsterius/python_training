```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(centers=3, random_state=42)
```

```python
X.shape
```

```python
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)

for (m1, m2), c in zip(model.coef_, model.intercept_):
    
    x1 = np.linspace(-10, 10, 50)
    x2 = (-x1*m1)/m2 - c/m2
    plt.ylim(-10, 10)
    plt.plot(x1, x2)
```

```python
colors = ["red", "yellow", "green", "black", "orange"]
```

```python
y_color = [colors[item] for item in y]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_color)

for index,((m1, m2), c) in enumerate(zip(model.coef_, model.intercept_)):
    x1 = np.linspace(-10, 10, 50)
    x2 = (-x1*m1)/m2 - c/m2
    plt.ylim(-10, 10)
    plt.plot(x1, x2, color=colors[index])
```
