```python
theta = np.linspace(0, 16*np.pi, 1000)
```

```python
r = np.linspace(5, 0, 1000) ** 2
```

```python
X = r * np.cos(theta)
```

```python
y = r * np.sin(theta)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, y)
```

```python
plt.scatter(X, y)
```

```python
X = np.linspace(-10, 10)
```

```python
plt.figure()
plt.subplot(3, 1, 1)
plt.plot(X, np.sin(X))

plt.subplot(3, 1, 2)
plt.plot(X, np.cos(X))

plt.subplot(3, 2, 5)
plt.plot(X, np.tan(X))

plt.subplot(3, 2, 6)
plt.plot(X, X**2)
plt.show()
```
